```
import matplotlib.pyplot as plt
import mtcnn
from mtcnn.mtcnn import MTCNN
from matplotlib.patches import Rectangle
```

```
pic = 'faces_det.jpg'
data = plt.imread(pic)
detector = MTCNN()
```

```
faces = detector.detect_faces(data)
```

```
# print(len(faces))
```

```
ax = plt.gca()
plt.imshow(data)
for face in faces:
	x, y, width, height = face['box']
	rect = plt.Rectangle((x, y), width, height, fill=False, color='green',)
	ax.add_patch(rect)

plt.show()
```

```
#Getting region of region of interest
for i in range(len(faces)):
    x, y, width, height = faces[i]['box']
    plt.subplot(len(faces),1,i+1)
    plt.axis('off')
    img = data[y:y+height, x:x+width]
    plt.imshow(img)
plt.show()
```
